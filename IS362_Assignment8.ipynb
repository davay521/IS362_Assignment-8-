### import all libraries and clean up data 

```python
%matplotlib inline
import pandas as pd 
import numpy as np 
import seaborn as sns
import sys
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
auto = pd.read_csv(url, sep = '\s+',names = ['MPG','Cylinders','Displacement','Horse power','Weight','Acceleration','Model Year','Origin','Car Name'])
auto.head(6)
```

### Convert '?' to NaN

```python
auto_data = auto.replace('?',np.nan)
auto_data.head(6)
```

### Convert Horse Power column to numeric value

```python
auto_data['Horse power'] = auto_data['Horse power'].astype(float)
auto_data.head(6)
```

### Change column values of origin to USA ASIA AND EUROPE

```python
auto_data['Origin'].replace({1:'USA',2:'ASIA',3:'EUROPE'},inplace=True)
auto_data.head(6)
```

### Create Bar plot using seaborn showing the distribution for cylinders 



```python
sns.set()

sns.barplot(y ='MPG',hue ='Cylinders',x= 'Origin',data = auto_data)
```

### Create a scatter plot showing the relationship between Horse power and weight 

```python
sns.lmplot(y = 'Weight',x = 'Horse power',data = auto_data)
```

## What is the relationship between Horse power and Accerleration?

```python
sns.lmplot(y = 'Horse power',x = 'Acceleration',data = auto_data)
```

### As horsepower of the vehicle decreases we see a Decrese in Acceleration
